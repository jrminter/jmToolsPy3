```python
from skimage import color, data
import matplotlib.pyplot as plt
from jmToolsPy3 import plotImageMontage

im1 = color.rgb2gray(data.astronaut())
im2 = data.coins()

myFig = plotImageMontage([im1, im2], ["Astronaut Eileen Collins","Roman Coins"], 1, 2,
                         title="Grayscale Montage", bare=True, cb=True);
```

```python
myFig = plotImageMontage([im1, im2], ["Astronaut Eileen Collins","Roman Coins"], 1, 2,
                         cmap = plt.cm.spectral, title="Spectral Montage", bare=True, cb=True);
```

```python
myFig = plotImageMontage([im1, im2], ["Astronaut Eileen Collins","Roman Coins"], 1, 2,
                         cmap = plt.cm.viridis, title="Viridis Montage", bare=True, cb=True);
```

```python
import skdemo
from skimage import data
import numpy as np

color_image = data.chelsea()

red_image = np.zeros_like(color_image)
green_image = np.zeros_like(color_image)
blue_image = np.zeros_like(color_image)

red_image[:, :, 0] = color_image[:, :, 0]
green_image[:, :, 1] = color_image[:, :, 1]
blue_image[:, :, 2] = color_image[:, :, 2]

skdemo.imshow_all(red_image, green_image, blue_image);

skdemo.strip_axis_labels();
green_image.dtype
```
